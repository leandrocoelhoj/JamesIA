```python
import requests

def treat_route(route_list):

    data = []
    for route in route_list:
        equipe = route['equipe'].split('-')[1].strip()
        techprefix = route['route'].split('Tech:')[1].split(')')[0].strip()
        nome_rota = route['route'].split('-')[1].split('(')[0].strip()

        data_info = {
            "equipe": equipe,
            "techprefix": techprefix,
            "nome_rota": nome_rota
        }
        data.append(data_info)
    return data

def post_routes(routes):
    r = requests.post(url="http://localhost:8000/api/ingest_routes/", json=routes)
    return r.status_code
```

```python
def get_api_callix_data(endpoint, domain, token, start_date, end_date):
    url = f"https://{domain}.callix.com.br/api/v1/{endpoint}"
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {token}"
    }
    querystring = {
        "filter[date]": f"{start_date}T00:00:00.000Z,{end_date}T23:59:59.999Z"
    }
    response = requests.get(url=url, headers=headers, params=querystring)
    response.raise_for_status()
    return response.json()
```

```python
def post_to_backend(endpoint, data, dominio):
    api_url = f"http://host.docker.internal:8000/api/callix_data/{endpoint}/"
    mapped_data = map_data(data, dominio)

    print(f"[API POST] Enviando {len(mapped_data)} registros para o endpoint {endpoint}")

    try:
        r = requests.post(api_url, json=mapped_data)
        try:
            resp_json = r.json()
        except Exception:
            resp_json = r.text
        if r.status_code >= 400:
            print(f"[API POST] Falha ao enviar dados do endpoint {endpoint}: {r.status_code} {resp_json}")
        else:
            print(f"[API POST] Sucesso ao enviar dados do endpoint {endpoint} para API. Status: {r.status_code}")
        r.raise_for_status()
        return resp_json
    except Exception as e:
        print(f"[API POST] Exceção ao enviar dados do endpoint {endpoint}: {e}")
        return None

def map_data(data, dominio):
    mapped = []
    # Se data é lista de dicts já
    if isinstance(data, list):
        for d in data:
            d['equipe'] = dominio
            if 'id' in d:
                d['operador_id'] = d['id']
            mapped.append(d)
        return mapped

    # Se é dict único sem "data"
    if isinstance(data, dict) and 'data' not in data:
        data['equipe'] = dominio
        if 'id' in data:
            data['operador_id'] = data['id']
        mapped.append(data)
        return mapped

    # Caso Callix original {"data": [...]}
    for item in data.get('data', []):
        attrs = item.get('attributes', {})
        attrs['equipe'] = dominio
        if 'id' in item:
            attrs['operador_id'] = item['id']
        mapped.append(attrs)
    return mapped
```

```python
def post_to_backend(endpoint, data, dominio):
    api_url = f"http://localhost:8000/api/callix_data/{endpoint}/"
    mapped_data = map_data(data=data, dominio=dominio, endpoint=endpoint)

    if not mapped_data:
        print(f"[API POST] payload vazio para {endpoint} | {dominio} — pulando POST.")
        return {"skipped": True, "count": 0}

    print(f"[API POST] Enviando {len(mapped_data)} registros para o endpoint {endpoint}.")

    r = requests.post(api_url, json=mapped_data)
    try:
        resp_body = r.json()
    except Exception:
        resp_body = r.text
    if r.status_code >= 400:
        print(f"[API POST] Falha ao enviar dados do endpoint {endpoint}: {r.status_code} {resp_body}")
    else:
        print(f"[API POST] Sucesso ao enviar dados do endpoint {endpoint} para API. Status: {r.status_code}")
    r.raise_for_status()
    return resp_body

def map_data(endpoint, data, dominio):
    mapped = []

    def attach_common(d):
        d['equipe'] = dominio
        return d

    if isinstance(data, list):
        for d in data:
            d = attach_common(d)
            if endpoint == 'user_performance_reports' and 'id' in d:
                d['operador_id'] = str(d['id'])
            mapped.append(d)
        return mapped

    # dict sem "data"
    if isinstance(data, dict) and 'data' not in data:
        data = attach_common(data)
        if endpoint == 'user_performance_reports' and 'id' in data:
            data['operador_id'] = str(data['id'])
        mapped.append(data)
        return mapped

    # {"data":[...]} (formato Callix)
    for item in data.get('data', []):
        attrs = (item.get('attributes') or {}).copy()
        attrs = attach_common(attrs)
        if endpoint == 'user_performance_reports' and 'id' in item:
            attrs['operador_id'] = str(item['id'])
        mapped.append(attrs)
    return mapped
```
